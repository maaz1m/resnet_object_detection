# MODEL IMPLEMENTATION:

```python
import os
import torch
import torch.nn as nn
import numpy as np

from torchvision.models import resnet18, ResNet18_Weights
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
class Model(nn.Module):

  def __init__(self):
    super(Model, self).__init__()
    pretrained_model = resnet18(weights=ResNet18_Weights.DEFAULT)
    self.backbone = nn.Sequential(*list(pretrained_model.children())[:-2])

    ### Initialize the required Layers
    self.pool = nn.AdaptiveAvgPool2d(output_size=(1, 1))
    self.have_object = nn.Linear(pretrained_model.fc.in_features, 2)
    self.cat_or_dog = nn.Linear(pretrained_model.fc.in_features, 3)
    self.specie = nn.Linear(pretrained_model.fc.in_features, 10)
    self.bbox = nn.Linear(pretrained_model.fc.in_features, 4)
    self.flatten = nn.Flatten()
    self.sigmoid = nn.Sigmoid()
    ### Initialize the required Layers

  def forward(self, input):
      x = self.backbone(input)
      x = self.pool(x)
      x = self.flatten(x)
      
      xobj = self.have_object(x)
      xcatdog = self.cat_or_dog(x)
      xspecie = self.specie(x)

      xbbox = self.bbox(x)
      xbbox = self.sigmoid(xbbox)

      return {
          "object": xobj,
          "cat_or_dog": xcatdog,
          "specie": xspecie,
          "bbox": xbbox
      }
```

# CUSTOM DATALOADER IMPLEMENTATION

```python
# class names we'll use in a lot of places

class_names = {
  'specie': ['NA', 'basset_hound','Birman','pomeranian','american_pit_bull_terrier','american_bulldog','Abyssinian','beagle','Persian','chihuahua'],
  'cat_or_dog': ['NA', 'cat', 'dog']
}
```

```python
train_list = list(np.load('/content/assessment_dataset/train_list.npy', allow_pickle=True).tolist())
val_list = list(np.load('/content/assessment_dataset/val_list.npy', allow_pickle=True).tolist())
```

```python
from bs4 import BeautifulSoup

def read_xml_file(path):
    with open(path, 'r') as f:
        data = f.read()
    bs_data = BeautifulSoup(data, 'xml')
    return {
        "object": 1,
        "cat_or_dog": bs_data.find("name").text,
        "xmin": int(bs_data.find("xmin").text),
        "ymin": int(bs_data.find("ymin").text),
        "xmax": int(bs_data.find("xmax").text),
        "ymax": int(bs_data.find("ymax").text),
        "specie": "_".join(path.split(os.sep)[-1].split("_")[:-1])
    }
```

```python
from PIL import Image, ImageDraw

import torchvision.transforms as transforms


class CustomDataset():

  def __init__(self, dataset_path, images_list, preprocess_flag=True, augment_flag=True):
    self.preprocess_flag = preprocess_flag
    self.augment_flag = augment_flag

    self.preprocess = transforms.Compose([
      transforms.Resize((224,224)),
      transforms.ToTensor(),
      transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),        
    ])
    self.augment = transforms.ColorJitter()

    image_folder_path = os.path.join(dataset_path, "images")
    label_folder_path = os.path.join(dataset_path, "labels")

    self.dataset = []
    for img_filename in os.listdir(image_folder_path):
        img_name = img_filename.split(os.sep)[-1].split(".")[0]
        if img_name in images_list:
          xml_path = os.path.join(label_folder_path, img_name+".xml")
          if os.path.exists(xml_path):
            xml_data = read_xml_file(xml_path)
          else:
            xml_data = {'object': 0, 'cat_or_dog': 'NA', 'specie': 'NA', 'xmin': 0, 'ymin': 0, 'xmax': 0, 'ymax': 0}
          xml_data['img_path'] = os.path.join(image_folder_path, img_filename)  
          self.dataset.append(xml_data)

  def __len__(self):
    return len(self.dataset)

  def __getitem__(self, index):
    data = self.dataset[index]
    image = Image.open(data['img_path'])
    width, height = image.size
    bbox = [data['xmin'] / width, data['ymin'] / height, data['xmax'] / width, data['ymax'] / height]
    if self.augment_flag:
      image = self.augment(image)
    if self.preprocess_flag:
      image = self.preprocess(image)

    label = {
        'object': data['object'],
        'cat_or_dog': class_names['cat_or_dog'].index(data['cat_or_dog']), # could be faster
        'specie': class_names['specie'].index(data['specie']),
        'bbox': torch.tensor(bbox)
    }
    return image, label
```

```python
dataset = CustomDataset('assessment_dataset', train_list, preprocess_flag=False, augment_flag=False)
for i in range(5):
  image, label = dataset[i]
  width, height = image.size
  draw = ImageDraw.Draw(image)
  draw.rectangle([(label['bbox'][0]*width, label['bbox'][1]*height), (label['bbox'][2]*width, label['bbox'][3]*height)])
  draw.text((0,0), class_names['cat_or_dog'][label['cat_or_dog']] + class_names['specie'][label['specie']])
  image.show()
```

# TRAINING LOOP IMPLEMENTATION

## Initializations

```python
from torch.utils.data import DataLoader
```

```python
# import our library
import torchmetrics

def train(epochs, checkpoint_path=None, best_model_path=None):

  # Initialize Model and Optimizer
  model = Model()
  model.to(device)
  optimizer = torch.optim.Adam(model.parameters(), lr=0.0001)
  epoch = 0

  if checkpoint_path and os.path.exists(checkpoint_path):
    checkpoint = torch.load(checkpoint_path, map_location=torch.device(device))
    model.load_state_dict(checkpoint['model_state_dict'])
    optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
    epoch = checkpoint['epoch'] + 1

  def save_model(path):
    torch.save({
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
    }, path)

  # Initialize Loss Functions
  loss_fn_dict = {
      'object': nn.CrossEntropyLoss(),
      'cat_or_dog': nn.CrossEntropyLoss(),
      'specie': nn.CrossEntropyLoss(),
      'bbox': nn.MSELoss()
  }

  training_dataset = CustomDataset("/content/assessment_dataset", images_list=train_list)
  training_loader = torch.utils.data.DataLoader(training_dataset, batch_size=32, shuffle=True)

  def train_one_epoch(epoch_index, tb_writer):
      running_loss = 0.
      last_loss = 0.

      # Here, we use enumerate(training_loader) instead of
      # iter(training_loader) so that we can track the batch
      # index and do some intra-epoch reporting
      for i, data in enumerate(training_loader):
          # Every data instance is an input + label pair
          inputs, labels = data
          inputs, labels = inputs.to(device), {key: labels[key].to(device) for key in labels}

          # zero the parameter gradients
          optimizer.zero_grad()

          # Make predictions for this batch
          outputs = model(inputs)

          # Compute the loss and its gradients
          loss_dict = {key: loss_fn_dict[key](outputs[key], labels[key]) for key in loss_fn_dict}

          loss =  sum(loss_dict.values())# Consolidate all individual losses
          
          loss.backward()
          optimizer.step()

          # Gather data and report
          running_loss += loss.item()
          if i % 10 == 0:
              last_loss = running_loss / 10 # loss per batch
              running_loss = 0.

      return last_loss

  model.train()
  
  best_metrics = {}

  while epoch < epochs:

    epoch_loss = train_one_epoch(i, None)
    print(f' Epoch {epoch} Loss : {epoch_loss}')

    save_model(checkpoint_path)

    curr_metrics = test(model, val_list=val_list)

    if sum(curr_metrics.values()) > sum(best_metrics.values()):
      best_metrics = curr_metrics
      save_model(best_model_path)
      print(f'Best model saved')

    print(curr_metrics)

    epoch+=1

def test(model, val_list):

  val_dataset = CustomDataset("/content/assessment_dataset", images_list=val_list, augment_flag=False)
  val_loader = torch.utils.data.DataLoader(val_dataset, batch_size=4, shuffle=True)

  metric_fn_dict = {
      'object': torchmetrics.F1Score(task='binary').to(device),
      'cat_or_dog': torchmetrics.F1Score(task='multiclass', num_classes=3).to(device),
      'specie': torchmetrics.F1Score(task='multiclass', num_classes=10).to(device)
  }

  model.eval()

  for i, data in enumerate(val_loader):
    inputs, labels = data
    inputs, labels = inputs.to(device), {key: labels[key].to(device) for key in labels}

    # Make predictions for this batch
    with torch.no_grad():
      outputs = model(inputs)

    # update metric function for each output key
    for key in metric_fn_dict:
      predictions = torch.max(outputs[key], 1)[1]
      metric_fn_dict[key].update(predictions, labels[key])


  score_dict = {key: metric_fn_dict[key].compute().item() for key in metric_fn_dict}
  return score_dict
```

```python
from PIL import ImageDraw

def logits_to_labels(logits):
  _, label = torch.max(logits, 1)
  return label


def visualize(model_weights_path, image_folder_path, output_folder="output"):

  model = Model()
  model.load_state_dict(torch.load(model_weights_path)['model_state_dict'])
  model.to(device)
  model.eval()
  if not os.path.exists(output_folder):
    os.makedirs(output_folder)

  preprocess = transforms.Compose([
        transforms.Resize((224,224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),        
  ])

  results = {}
  for img_fname in os.listdir(image_folder_path)[:10]:
    img_path = os.path.join(image_folder_path, img_fname)
    image = Image.open(img_path)
    width, height = image.size

    image_preprocessed = preprocess(image)
    image_batched = image_preprocessed.unsqueeze(0)
    image_batched = image_batched.to(device)

    with torch.no_grad():
      outputs = model(image_batched)

    scale = torch.tensor([[width,height,width,height]])
    xmin, ymin, xmax, ymax = (outputs['bbox'].cpu()*scale).squeeze().to(torch.int64).tolist()

    results[img_fname] = {
      'xmin': xmin,
      'ymin': ymin,
      'xmax': xmax,
      'ymax': ymax,
      'object': logits_to_labels(outputs['object']).item(),
      'cat_or_dog': class_names['cat_or_dog'][logits_to_labels(outputs['cat_or_dog']).item()],
      'specie': class_names['specie'][logits_to_labels(outputs['specie']).item()]
    }
    draw = ImageDraw.Draw(image)
    draw.rectangle([(xmin,ymin),(xmax,ymax)])
    image.save(os.path.join(output_folder, img_fname))
    
  return results
```
